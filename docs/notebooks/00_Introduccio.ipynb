# Benvinguda

Aquest curs de Machine Learning avançat en Python per a Aplicacions en Oceanografia i Clima ofereix una formació pràctica i aplicada en tècniques modernes d’aprenentatge automàtic, adaptades a problemes reals en contextos científics.

Es treballarà amb models de Machine Learning supervisat i no supervisat utilitzant scikit-learn, incloent SVM, models lineals i mètodes d’ensembles com bagging i boosting. Es desenvoluparan xarxes neuronals amb PyTorch y Scikit-learn, des de Multi Layer Perceptron fins a xarxes convolucionals per a l’anàlisi d’imatges, i s’abordarà el tractament de dades temporals específiques de l’oceà i el clima.

A més, el curs inclou tècniques per al processament de text científic, des de classificació fins a generació de contingut amb embeddings i arquitectures RAG, permetent als estudiants integrar dades heterogènies en projectes avançats de recerca i aplicacions pràctiques.
