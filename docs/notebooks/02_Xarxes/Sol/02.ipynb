Dataset description

[Ocean Quality](https://www.kaggle.com/datasets/vinothkannaece/ocean-quality-dataset)

This dataset is a synthetically generated ocean water quality dataset designed for machine learning experiments related to environmental monitoring, water quality prediction, and classification tasks. It simulates realistic oceanic conditions with continuous and categorical features, incorporating probabilistic variation and random noise to prevent model overfitting.

The dataset consists of 100,000 records and 8 features, including physicochemical parameters commonly used in ocean quality assessment


1. Given the *Ocean Quality* dataset, train an MLP to predict the value of variable *Quality*.
2. Find the best MLP to predict the Salinity. Exclude the Quality from the dataset for this exercise.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import mean_absolute_error, mean_squared_error

random_state=33
```

## 1

```python
df = pd.read_csv("ocean/synthetic_ocean_quality_realistic.csv")
print(df.columns)
```

```python
df["Quality"].value_counts()
```

```python
X = np.array(df.iloc[:,0:7])
le = LabelEncoder()
y = le.fit_transform(df["Quality"])
```

```python
# Split into train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)
```

```python
# Normalization
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Training
model = MLPClassifier(
    hidden_layer_sizes=(32, 8),
    activation='relu',
    solver='adam',
    learning_rate_init=0.001,
    max_iter=1000,
    random_state=random_state,
)
model.fit(X_train, y_train)
```

```python
y_pred_train = model.predict(X_train)
print(classification_report(y_train, y_pred_train))
```

```python
# Prediction and evaluation
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

Preparam les dades

```python
y = np.array(df["Salinity"])

df2 = df.drop(["Salinity", "Quality"], axis=1)
X = np.array(df2)

print(X.shape, y.shape)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)

scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)
```

 Regressió amb MLPRregressor

```python

from sklearn.neural_network import MLPRegressor

model = MLPRegressor(
    hidden_layer_sizes=(100,100),
    activation='relu',
    solver='adam',
    batch_size='auto',
    learning_rate='constant',
    learning_rate_init=0.1,
    shuffle=True,
    random_state=random_state
)
```

```python
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print(mean_absolute_error(y_test, y_pred))
```

```python
model.fit(X_train_s, y_train)

y_pred = model.predict(X_test_s)
print(mean_absolute_error(y_test, y_pred))
```

```python

# Model
model = MLPRegressor(
    hidden_layer_sizes=(64, 32),
    activation='relu',
    solver='adam',
    learning_rate_init=0.001,
    max_iter=500,
    random_state=42
)
model.fit(X_train_s, y_train)

# Avaluació
y_pred = model.predict(X_test_s)
mae  = mean_absolute_error(y_test, y_pred)

print(f"MAE:  {mae:.4f}")
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred)) # evitar grans errors
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100

print(f"RMSE: {rmse:.4f}")
print(f"MAPE: {mape:.2f}%")
```

GridSearch

The application of Grid search over a regression problem is tricky:

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'hidden_layer_sizes': [(64, 32), (64, 32, 16), (64,), (32,), (16,), (100,)],
    'learning_rate_init': [0.001, 0.01],
    'max_iter': [500]
}

mlp_regressor = MLPRegressor(random_state=42, solver='adam')
grid_search = GridSearchCV(
    estimator=mlp_regressor,
    param_grid=param_grid,
    cv=5,
    scoring='neg_mean_squared_error', # here!
    n_jobs=-1,
    verbose=1
)

grid_search.fit(X_train, y_train)

print(f"Best parameters: {grid_search.best_params_}")
print(f"Best RMSE: {(-grid_search.best_score_) ** 0.5:.4f}")

best_model = grid_search.best_estimator_
```
