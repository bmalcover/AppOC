## Projecte: Classificació amb el dataset CalCOFI

El dataset CalCOFI (California Cooperative Oceanic Fisheries Investigations) conté mesures oceanogràfiques recollides a la costa de Califòrnia des de 1949, incloent variables com temperatura, salinitat, oxigen dissolt, fosfats i clorofil·la, entre d'altres. En aquest projecte aplicareu les tècniques apreses als blocs 1 i 2 per construir i comparar diferents models de classificació.

Aquest projecte té els següents objectius:

- Aplicar i comparar classificadors clàssics de ML amb un MLP Cercar la millor configuració de cada model de manera sistemàtica.
- Implementar el mateix problema tant amb scikit-learn com amb PyTorch.
- Avaluar els models de manera rigorosa i reproducible.

Descarregueu el dataset CalCOFI de Kaggle. Definireu vosaltres mateixos la variable objectiu de classificació a partir de les variables disponibles (per exemple, classificació de masses d'aigua per rangs de temperatura, salinitat o oxigen dissolt). Haureu de justificar la vostra elecció.

Heu d'acomplir les següents tasques:

1. Preparació de les dades:
- Càrrega i exploració inicial del dataset.
- Divisió en conjunts d'entrenament i test (80/20). Aquesta partició ha de ser la mateixa per a tots els experiments del projecte.
- Normalització de les dades.

2. Classificadors amb scikit-learn:
- Empreu com a mínim tres dels classificadors vists en el bloc 1. Un d'ells ha de ser _Random Forest_.
- Per cada classificador. Cerqueu la millor configuració d'hiperparàmetres de manera automàtica emprant GridSearchCV o RandomizedSearchCV amb k-folding sobre el conjunt d'entrenament Avalueu el millor model obtingut sobre el conjunt de test
- Discutiu les mètriques d'avaluació

3. MLP amb scikit-learn

- Implementeu un MLPClassifier i cerqueu la millor configuració d'hiperparàmetres (arquitectura, funció d'activació, learning rate) emprant k-folding. Avalueu el millor model sobre el conjunt de test

4. MLP amb PyTorch

- Implementeu un MLP equivalent al de scikit-learn emprant nn.Sequential.
- Entreneu el model i guardeu els pesos.
- Genereu un bloc de codi independent que carregui el model guardat i avaluï el conjunt de test sense tornar a entrenar.

5. Comparació i conclusions

- Construïu una taula comparativa amb les mètriques de tots els models (accuracy, precision, recall, F1-score)
- Discutiu quins models han funcionat millor i per quines raons
