## Project: Classificació amb el dataset CalCOFI

The CalCOFI dataset (California Cooperative Oceanic Fisheries Investigations) contains oceanographic measurements collected off the coast of California since 1949, including variables such as temperature, salinity, dissolved oxygen, phosphates, and chlorophyll, among others. In this project you will apply the techniques learned in blocks 1 and 2 to build and compare different classification models.

This project has the following objectives:

- Apply and compare classic ML classifiers with an MLP. Find the best configuration for each model systematically.
- Implement the same problem with both scikit-learn and PyTorch.
- Evaluate models rigorously and reproducibly.

Download the CalCOFI dataset from Kaggle. You will define the classification target variable from the available variables (for example, water mass classification by temperature ranges, salinity, or dissolved oxygen). You must justify your choice.

You must complete the following tasks:

1. Data Preparation:
- Load and initial exploration of the dataset.
- Split into training and test sets (80/20). This partition must be the same for all project experiments.
- Data normalization.

2. Classifiers with scikit-learn:
- Use at least three of the classifiers seen in block 1. One of them must be _Random Forest_.
- For each classifier. Find the best hyperparameter configuration automatically using GridSearchCV or RandomizedSearchCV with k-folding on the training set. Evaluate the best model obtained on the test set
- Discuss the evaluation metrics

3. MLP with scikit-learn

- Implement an MLPClassifier and find the best hyperparameter configuration (architecture, activation function, learning rate) using k-folding. Evaluate the best model on the test set

4. MLP with PyTorch

- Implement an MLP equivalent to scikit-learn using nn.Sequential.
- Train the model and save the weights.
- Generate an independent block of code that loads the saved model and evaluates the test set without retraining.

5. Comparison and Conclusions

- Build a comparative table with metrics from all models (accuracy, precision, recall, F1-score)
- Discuss which models performed best and why
